```python
dates = [ 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31 ]
days = { '일': 0, '월': 1, '화': 2, '수': 3, '목': 4, '금': 5, '토': 6 }
key_days = list(days.keys())
period_day = [ 1, 1, 8, 8, 15, 15, 22, 22, 29, 29 ]
start_days = [ '화', '금', '금', '월', '수', '토', '월', '목', '일', '화', '금', '일' ]
start_day = days['화']

def blank(date):
    if date < 10:
        return f'{date}   '
    else:
        return f'{date}  '
    
def start_blank(start_day):
    for start_blank in range(7):
        if start_day == 0:
            return ''
        elif start_day == start_blank:
            return '    ' * start_blank

def period(*args):
    return 7 - int(start_day) + period_day[int(*args)]      
       

for mon in range(1, 13):
    print(mon, '월')

    for day in range(7):
        print(key_days[day], end='  ')
    print('\n---------------------------')
    print(start_blank(start_day), end='')
    for fir_week in range(1, period(0)):
        print(blank(fir_week), end='')
    print()
    for num in range(1, 9, 2):
        for stf_week in range(period(num), period(num+1)):
            print(blank(stf_week), end='')
        print()
    if len(range(period(7), dates[mon-1] + 1)) > 7:
        for six_week in range(period(8), dates[mon-1] + 1):
            print(blank(six_week), end='')
    else:
        for fif_week in range(period(7), dates[mon-1] + 1):
                print(blank(fif_week), end='')

    print('\n')


    
# def final_date(start_day):
#     if start_day == 5 or 6:
#         return len(range(period(9), dates[mon-1]))
#     else:
#         return len(range(period(7), dates[mon-1]))

# def next_start_day(final_date):
#     if final_date(start_day) == 6:
#         return 0
#     else:
#         return final_date(final_date) + 1
```
