```python
import csv
```

```python
f_w = open('ss3.csv', 'a+', encoding='utf-8', newline='')

writer = csv.writer(f_w)
for i in range(7):
    writer.writerow(
    [i, i, i, i]
    )
f_w.close()
```

```python
f_r = open('ss3.csv', 'r', encoding='utf-8', newline='')

reader = csv.reader(f_r)

for line in reader:
    for e in line:
        print(e) 

#line은 list 한줄한줄이 다 리스트, 엔터마져 리스트로 만들어버림...
```
